```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array([1, 2, 3, 4, 5])  # np.array로 List 객체 변환
mean = arr.mean()  # 평균 구하기
std = arr.std()  # 표준편차 구하기
vfunc = np.vectorize(lambda x: (x - mean) / std)  # vectorize method를 통해 array를 변환 시켜주는 함수
new_arr = vfunc(arr)  # 정규화 된 array, 결과 값으로는 array([-1.41421356, -0.70710678,  0,  0.70710678,  1.41421356])

mean = 0
std = 1
variance = np.square(std)
x = np.arange(-3,3,.01)  # x축 데이터: -3 ~ 3 사이 0.01 간격 [-3, -2.99 ... 2.99, 3]
f = np.exp(-np.square(x - mean)/ 2 * variance)/(np.sqrt(2 * np.pi * variance))  # y축 데이터

plt.plot(x,f)  # x 좌표, y 좌표에 해당하는 데이터를 통해 그래프 표현
plt.show()
```
